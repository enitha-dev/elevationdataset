```python
import pandas as pd  
import matplotlib.pyplot as plt
import seaborn as sns

# Load residuals
m1 = pd.read_csv("resi_svm.csv")  
m2 = pd.read_csv("resi_rf.csv")
m3 = pd.read_csv("resi_knn.csv")  
m4 = pd.read_csv("resi_dt.csv")  

# Create figure  
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12,8))

# Histogram plots
ax1.hist(m1, bins=40)  
ax2.hist(m2, bins=40)
ax3.hist(m3, bins=40)
ax4.hist(m4, bins=40)

# Box plots   
sns.boxplot(data=m1, ax=ax1) 
sns.boxplot(data=m2, ax=ax2)
sns.boxplot(data=m3, ax=ax3)  
sns.boxplot(data=m4, ax=ax4)

ax1.set(xlabel='SVM Residuals', ylabel='Frequency')
ax2.set(xlabel='RF Residuals', ylabel='Frequency') 
ax3.set(xlabel='KNN Residuals', ylabel='Frequency')
ax4.set(xlabel='DT Residuals', ylabel='Frequency')

plt.tight_layout()
plt.show()
```
