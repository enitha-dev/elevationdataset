```python
import numpy as np
import matplotlib.pyplot as plt
r2_1 = float(np.loadtxt('dt_r2.txt')) 
r2_2 = float(np.loadtxt('knn_r2.txt'))
r2_3 = float(np.loadtxt('rf_r2.txt'))
r2_4 = float(np.loadtxt('svm_r2.txt'))
print('decision tree R^2: ', r2_1) 
print('knn R^2: ', r2_2)
print('random forest 3 R^2: ', r2_3)
print('svm R^2: ', r2_4)
```

```python
models = ["dt","knn","rf","svm"]
r2_scores = [abs(r2_1), r2_2, r2_3,r2_4]
best_model = np.argmax(r2_scores)

plt.bar([0,1,2,3], r2_scores, color=['blue','yellow','red','grey']) 
plt.bar(best_model, r2_scores[best_model], color='green', label='Best Model')
plt.legend()
plt.ylabel('Adjusted R^2')
plt.xlabel(models)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

r2_1 = float(np.loadtxt('dt_r2.txt')) 
r2_2 = float(np.loadtxt('knn_r2.txt'))
r2_3 = float(np.loadtxt('rf_r2.txt'))
r2_4 = float(np.loadtxt('svm_r2.txt'))

print('decision tree R^2: ', r2_1) 
print('knn R^2: ', r2_2)
print('random forest 3 R^2: ', r2_3)
print('svm R^2: ', r2_4)

models = ["dt","knn","rf","svm"]
r2_scores = [abs(r2_1), r2_2, r2_3, r2_4]
best_model_index = np.argmax(r2_scores)
best_model = models[best_model_index]

plt.plot(models, r2_scores, marker='o', color='yellow', label='R^2 Scores')
plt.plot(best_model, r2_scores[best_model_index], marker='o', color='red', label='Best Model')
plt.legend()
plt.ylabel('Adjusted R^2')
plt.xlabel('Models')
plt.title('Adjusted R^2 Scores for Different Models')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load R^2 scores for each model
r2_1 = float(np.loadtxt('dt_r2.txt')) 
r2_2 = float(np.loadtxt('knn_r2.txt'))
r2_3 = float(np.loadtxt('rf_r2.txt'))
r2_4 = float(np.loadtxt('svm_r2.txt'))

# Print R^2 scores
print('Decision Tree R^2: ', r2_1) 
print('KNN R^2: ', r2_2)
print('Random Forest R^2: ', r2_3)
print('SVM R^2: ', r2_4)

# Model names and R^2 scores
models = ["Decision Tree", "KNN", "Random Forest", "SVM"]
r2_scores = [r2_1, r2_2, r2_3, r2_4]

# Plot each model's adjusted R^2 scores as separate lines
for model, r2_score in zip(models, r2_scores):
    plt.plot([model], [r2_score], marker='o', label=model)

plt.legend()
plt.ylabel('Adjusted R^2')
plt.xlabel('Models')
plt.title('Adjusted R^2 Scores for Different Models')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.grid(True)
plt.show()
```
